```r
#  Paso 2: elegir el tamaño del árbol que minimice la tasa de datos mal clasificados (i.e. error de predicción).
# Buscamos el valor de cp que minimice xerror con el árbol mas simple.

bestcp_all_gini_32 <- modelo1_all_maximal_gini_32$cptable[which.min(modelo1_all_maximal_gini_32$cptable[,"xerror"]),"CP"]

# Paso 3: Podar el árbol maximal usando el mejor cp.
modelo1_all_pruned_gini_32 <- prune(modelo1_all_maximal_gini_32, cp = bestcp_all_gini_32)
```
